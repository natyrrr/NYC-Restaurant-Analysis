# **Phase 2 Restaurant Opening Applications**

This data will be another resource we will use to analyze the amount of restaurants that have closed. We will compare the amount of restaurants in this data to the amount of restaurants that were active during this same time last year. We will use this along with the eater data, to illustrates restaurants that did not make it. All data will be compared to this same time last year. The idea is that if they have an application they survived the pandemic at this point in time.

```python
# Import necessary libraries and dependencies.
import pandas as pd
import numpy as np
from pathlib import Path
import csv
import sqlalchemy 
from sqlalchemy import create_engine
```

```python
csv_path=Path("open_restaurant_applications.csv")
```

```python
applications_df=pd.read_csv(csv_path)
```

```python
applications_df.head()
```

```python
open_restaurants=applications_df.drop(columns=["globalid","Community Board","Council District","Census Tract","BIN","BBL"])
```

```python
open_restaurants=open_restaurants.drop(columns=["Qualify Alcohol","SLA Serial Number","SLA License Type","Landmark District or Building","landmarkDistrict_terms","healthCompliance_terms"])
```

```python
open_restaurants=open_restaurants.drop(columns=["Sidewalk Dimensions (Area)","Roadway Dimensions (Length)","Roadway Dimensions (Width)","Roadway Dimensions (Area)","Approved for Sidewalk Seating","Approved for Roadway Seating"])
```

```python
open_restaurants=open_restaurants.drop(columns=["Food Service Establishment Permit #", "Sidewalk Dimensions (Length)", "Sidewalk Dimensions (Width)"])
```

```python
open_restaurants=open_restaurants.rename(columns={"Seating Interest Sidewalk-Roadway-Both": "Seating Interest"})
```

```python
open_restaurants.head()
```

```python
# Create New CSV
open_restaurants.to_csv("clean_open_restaurants.csv")
```
